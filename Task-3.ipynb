```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [23, 45, 56, 78],
    'Growth': [5, 9, 7, 3]
}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 4))
plt.bar(df['Category'], df['Values'], color='blue', label='Values')
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Bar Chart of Values by Category')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(df['Category'], df['Growth'], color='green', marker='o', label='Growth')
plt.xlabel('Category')
plt.ylabel('Growth')
plt.title('Line Chart of Growth by Category')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Category'], df['Values'], color='blue', label='Values')
plt.plot(df['Category'], df['Growth'], color='green', marker='o', label='Growth')
plt.xlabel('Category')
plt.ylabel('Values / Growth')
plt.title('Combined Bar and Line Chart')
plt.legend()
plt.show()
```
