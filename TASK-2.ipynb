```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'INDIAvi.csv' 
data = pd.read_csv(r"C:\Users\HP\Desktop\Python\INDIAvi.csv")
print(data.info())
print(data.head())
```

```python
summary_statistics = data.describe()
print(summary_statistics)
```

```python
plt.figure(figsize=(14, 10))
plt.subplot(2, 2, 1)
sns.histplot(data['views'], kde=True)
plt.title('Distribution of Views')
plt.subplot(2, 2, 2)
sns.histplot(data['likes'], kde=True)
plt.title('Distribution of Likes')
plt.subplot(2, 2, 3)
sns.histplot(data['dislikes'], kde=True)
plt.title('Distribution of Dislikes')
plt.subplot(2, 2, 4)
sns.histplot(data['comment_count'], kde=True)
plt.title('Distribution of Comment Count')
plt.tight_layout()
plt.show()
```
